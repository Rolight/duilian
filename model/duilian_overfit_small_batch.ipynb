```python
import duilian
```

```python
# overfit small batch
model = duilian.DuiLianModel(batch_size=5, learning_rate=0.03, l2_reg=1e-5, use_pretrained_embedding=False)
model.init_variables()
batch = model.generate_train_batch()
batch_inputs, batch_inputs_length = batch[0], batch[1]
batch_outputs = batch[4]
print('training data output:')
model.print_sample_id(batch_outputs)
```

```python
# overfit small batch
for i in range(500):
    loss, _ = model.training_on_batch(model.sess, *batch, 1)
    if i % 20 == 0:
        print('on epoch %d, loss is %.10f' % (i, loss))
        model.predict(batch_inputs, batch_inputs_length)
```
